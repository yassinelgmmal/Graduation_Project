```python
import numpy as np
from PIL import Image
from paddleocr import PaddleOCR

# Process each detected section
for i, block in enumerate(layout):
    x1, y1, x2, y2 = map(int, block.coordinates)  # Get bounding box

    # Crop detected region from the clean image
    cropped_img = image_for_ocr[y1:y2, x1:x2]

    # Convert cropped image back to PIL format for OCR
    pil_img = Image.fromarray(cropped_img)

    # Perform OCR using PaddleOCR
    result = ocr.ocr(np.array(pil_img), cls=True)

    # Extract and format text
    extracted_text = "\n".join([line[1][0] for line in result[0] if line[1][0].strip()])

    # Print structured output
    print(f"Section {i+1} ({block.type} - Confidence: {block.score:.2f}):\n{extracted_text}\n")
```
